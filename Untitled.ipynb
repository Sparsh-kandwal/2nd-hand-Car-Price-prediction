```python
final_dataset['no_year']=final_dataset['Current_year']-final_dataset['Year']
```

```python
final_dataset.drop(['Year'],axis=1,inplace=True)
```

```python
final_dataset.drop(['Current_year'],axis=1,inplace=True)
```
